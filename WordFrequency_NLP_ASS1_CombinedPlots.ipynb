```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python

document_text = open('/content/gdrive/MyDrive/Colab Notebooks/sinhala_large.txt', 'r')
corpus = document_text.read()
print(corpus)
```

```python
corpus_tokens = word_tokenize(corpus)
print(corpus_tokens)
```

```python
from nltk.util import ngrams
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.collocations import *

import matplotlib.pyplot as plt
import numpy as np
```

```python
bigrams = ngrams(corpus_tokens, 2)

bigrams_c = {}
for b in bigrams:
    if b not in bigrams_c:
        bigrams_c[b] = 1
    else:
        bigrams_c[b] += 1

import operator
sorted_bigrams = dict( sorted(bigrams_c.items(), key=operator.itemgetter(1),reverse=True))
print('Dictionary in descending order by value : ',sorted_bigrams.values())

values_view = sorted_bigrams.values()
value_iterator = iter(values_view)
bigram_first_value = next(value_iterator)


bigram_rank_dict ={"Rank":[],"Freq":[]}
bigram_rank = 1
highest = bigram_first_value
for b in sorted_bigrams.values():
  if b == highest:
    bigram_rank_dict["Rank"].append(bigram_rank)
  else:
    bigram_rank=bigram_rank+1
    bigram_rank_dict["Rank"].append(bigram_rank)
    highest = b

  bigram_rank_dict["Freq"].append(b)


print(bigram_rank_dict)
```

```python
trigrams = ngrams(corpus_tokens, 3)

trigrams_c = {}
for b in trigrams:
    if b not in trigrams_c:
        trigrams_c[b] = 1
    else:
        trigrams_c[b] += 1

import operator
sorted_trigrams = dict( sorted(trigrams_c.items(), key=operator.itemgetter(1),reverse=True))
print('Dictionary in descending order by value : ',sorted_trigrams.values())

values_view = sorted_trigrams.values()
value_iterator = iter(values_view)
trigram_first_value = next(value_iterator)

trigram_rank_dict ={"Rank":[],"Freq":[]}
trigram_rank = 1
highest = trigram_first_value
for b in sorted_trigrams.values():
  if b == highest:
    trigram_rank_dict["Rank"].append(trigram_rank)
  else:
    trigram_rank=trigram_rank+1
    trigram_rank_dict["Rank"].append(trigram_rank)
    highest = b

  trigram_rank_dict["Freq"].append(b)


print(trigram_rank_dict)
```

```python
fourgrams = ngrams(corpus_tokens, 4)

fourgrams_c = {}
for b in fourgrams:
    if b not in fourgrams_c:
        fourgrams_c[b] = 1
    else:
        fourgrams_c[b] += 1

import operator
sorted_fourgrams = dict( sorted(fourgrams_c.items(), key=operator.itemgetter(1),reverse=True))
print('Dictionary in descending order by value : ',sorted_fourgrams.values())

values_view = sorted_fourgrams.values()
value_iterator = iter(values_view)
fourgram_first_value = next(value_iterator)

fourgram_rank_dict ={"Rank":[],"Freq":[]}
fourgram_rank = 1
highest = fourgram_first_value
for b in sorted_fourgrams.values():
  if b == highest:
    fourgram_rank_dict["Rank"].append(fourgram_rank)
  else:
    fourgram_rank=fourgram_rank+1
    fourgram_rank_dict["Rank"].append(fourgram_rank)
    highest = b

  fourgram_rank_dict["Freq"].append(b)


print(fourgram_rank_dict)
```

```python
xpoints = np.array(bigram_rank_dict["Rank"])
ypoints = np.array(bigram_rank_dict["Freq"])
plt.xlabel("Rank")
plt.ylabel("Frequency")
plt.plot(xpoints, ypoints)
plt.title("Bigram Freq-Rank Plot")
plt.show()

xpoints = np.array(trigram_rank_dict["Rank"])
ypoints = np.array(trigram_rank_dict["Freq"])
plt.xlabel("Rank")
plt.ylabel("Frequency")
plt.plot(xpoints, ypoints)
plt.title("Trigram Freq-Rank Plot")
plt.show()


xpoints = np.array(fourgram_rank_dict["Rank"])
ypoints = np.array(fourgram_rank_dict["Freq"])
plt.xlabel("Rank")
plt.ylabel("Frequency")
plt.plot(xpoints, ypoints)
plt.title("Fourgram Freq-Rank Plot")
plt.show()
```
